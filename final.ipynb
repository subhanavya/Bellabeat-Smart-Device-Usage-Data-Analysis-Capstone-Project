```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')

# Load datasets (update filenames/paths accordingly)
daily_activity = pd.read_csv('dailyActivity_merged.csv')
hourly_calories = pd.read_csv('hourlyCalories_merged.csv')
hourly_intensities = pd.read_csv('hourlyIntensities_merged.csv')
hourly_steps = pd.read_csv('hourlySteps_merged.csv')
heartrate_seconds = pd.read_csv('heartrate_seconds_merged.csv')
minute_calories = pd.read_csv('minuteCaloriesNarrow_merged.csv')
minute_intensities = pd.read_csv('minuteIntensitiesNarrow_merged.csv')
minute_mets = pd.read_csv('minuteMETsNarrow_merged.csv')
minute_sleep = pd.read_csv('minuteSleep_merged.csv')
minute_steps = pd.read_csv('minuteStepsNarrow_merged.csv')
weight_loginfo = pd.read_csv('weightLogInfo_merged.csv')






def clean_inspect(df, date_col=None):
    print(f"--- Inspecting {df.name if hasattr(df, 'name') else 'Dataframe'} ---")
    print("Shape:", df.shape)
    print(df.head())
    print(df.info())
    print("Missing values:\n", df.isnull().sum())
    print("Duplicate rows:", df.duplicated().sum())
    df.drop_duplicates(inplace=True)
    if date_col:
        df[date_col] = pd.to_datetime(df[date_col])
    print("\n")



def clean_inspect(df, date_cols=None):
    print(f"--- Inspecting {df.name if hasattr(df, 'name') else 'Dataframe'} ---")
    print("Shape:", df.shape)
    print(df.head())
    print(df.info())
    print("Missing values:\n", df.isnull().sum())
    print("Duplicate rows:", df.duplicated().sum())
    df.drop_duplicates(inplace=True)
    
    if date_cols:
        if isinstance(date_cols, str):
            date_cols = [date_cols]
        for col in date_cols:
            if col in df.columns:
                df[col] = pd.to_datetime(df[col])
            else:
                print(f"Warning: Column '{col}' not found in dataframe.")
    print("\n")


clean_inspect(daily_activity, 'ActivityDate')
clean_inspect(hourly_calories, 'ActivityHour')
clean_inspect(hourly_intensities, 'ActivityHour')
clean_inspect(hourly_steps, 'ActivityHour')
clean_inspect(heartrate_seconds, 'Time')
clean_inspect(minute_calories, 'ActivityMinute')
clean_inspect(minute_intensities, 'ActivityMinute')
clean_inspect(minute_mets, 'ActivityMinute')

# Check actual columns before for minute_sleep and adjust accordingly:
print(minute_sleep.columns)
# Assuming the correct column name for datetime is different or doesn't exist:
# You can call clean_inspect without date conversion or with a correct column if found
clean_inspect(minute_sleep)  # no date_col passed

clean_inspect(minute_steps, 'ActivityMinute')
clean_inspect(weight_loginfo, 'Date')

# Assign names for printing
daily_activity.name = "Daily Activity"
hourly_calories.name = "Hourly Calories"
hourly_intensities.name = "Hourly Intensities"
hourly_steps.name = "Hourly Steps"
heartrate_seconds.name = "Heartrate Seconds"
minute_calories.name = "Minute Calories"
minute_intensities.name = "Minute Intensities"
minute_mets.name = "Minute METs"
minute_sleep.name = "Minute Sleep"
minute_steps.name = "Minute Steps"
weight_loginfo.name = "Weight Log Info"

# Perform cleaning and inspection on main datasets
clean_inspect(daily_activity, 'ActivityDate')
clean_inspect(hourly_calories, 'ActivityHour')
clean_inspect(hourly_intensities, 'ActivityHour')
clean_inspect(hourly_steps, 'ActivityHour')
clean_inspect(heartrate_seconds, 'Time')
clean_inspect(minute_calories, 'ActivityMinute')
clean_inspect(minute_intensities, 'ActivityMinute')
clean_inspect(minute_mets, 'ActivityMinute')
clean_inspect(minute_sleep, 'ActivityMinute')
clean_inspect(minute_steps, 'ActivityMinute')
clean_inspect(weight_loginfo, 'Date')

# Exploratory Data Analysis (EDA) 
# Focused EDA on daily_activity dataset as main summary table for Leaf product users

# Descriptive stats on key columns
print("Summary statistics for Daily Activity:")
print(daily_activity.describe())

# Number of unique users and dates
unique_users = daily_activity['Id'].nunique()
unique_dates = daily_activity['ActivityDate'].nunique()
print(f"Unique users: {unique_users}, Unique days tracked: {unique_dates}")

# Distribution of Total Steps
plt.figure(figsize=(10,5))
sns.histplot(daily_activity['TotalSteps'], bins=30, kde=True)
plt.title('Distribution of Daily Steps')
plt.xlabel('Steps')
plt.ylabel('Frequency')
plt.show()

# Scatter plot Calories vs Total Steps
plt.figure(figsize=(10,5))
sns.scatterplot(data=daily_activity, x='TotalSteps', y='Calories')
plt.title('Calories Burned vs Total Steps')
plt.xlabel('Total Steps')
plt.ylabel('Calories Burned')
plt.show()

# Boxplot Steps by weekday
daily_activity['Weekday'] = daily_activity['ActivityDate'].dt.day_name()
plt.figure(figsize=(10,5))
sns.boxplot(data=daily_activity, x='Weekday', y='TotalSteps', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Daily Steps by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Total Steps')
plt.show()

# Correlation matrix heatmap of select variables
plt.figure(figsize=(8,6))
corr = daily_activity[['TotalSteps', 'Calories', 'VeryActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Activity Metrics')
plt.show()

# ========== Advanced Analysis ==========

# 1. Identify most active users (top 10% by avg daily steps)
user_avg_steps = daily_activity.groupby('Id')['TotalSteps'].mean().sort_values(ascending=False)
top_10_percent_cutoff = user_avg_steps.quantile(0.90)
most_active_users = user_avg_steps[user_avg_steps >= top_10_percent_cutoff]
print("Top 10% most active users (by average daily steps):")
print(most_active_users)

# 2. Activity level clusters using k-means clustering on daily activity features (TotalSteps and VeryActiveMinutes)
from sklearn.cluster import KMeans
features = daily_activity[['TotalSteps', 'VeryActiveMinutes']].fillna(0)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(features)
daily_activity['ActivityCluster'] = clusters

# Visualize clusters
plt.figure(figsize=(10,6))
sns.scatterplot(data=daily_activity, x='TotalSteps', y='VeryActiveMinutes', hue='ActivityCluster', palette='Set1')
plt.title('User Activity Clusters')
plt.xlabel('Total Steps')
plt.ylabel('Very Active Minutes')
plt.show()

# Cluster insights: mean values per cluster
cluster_summary = daily_activity.groupby('ActivityCluster')[['TotalSteps', 'VeryActiveMinutes', 'Calories']].mean()
print("Cluster Summary:")
print(cluster_summary)

# 3. Sleep vs Activity analysis (if sleep data available)
if 'SleepMinutes' in minute_sleep.columns:
    sleep_summary = minute_sleep.groupby('Id').agg({'SleepMinutes': 'mean'}).rename(columns={'SleepMinutes': 'AvgSleepMinutes'})
    activity_summary = daily_activity.groupby('Id').agg({'TotalSteps': 'mean', 'Calories': 'mean'}).rename(columns={'TotalSteps': 'AvgSteps', 'Calories': 'AvgCalories'})
    sleep_activity = sleep_summary.join(activity_summary, how='inner')
    
    # Scatter plot sleep vs avg steps
    plt.figure(figsize=(10,6))
    sns.scatterplot(data=sleep_activity, x='AvgSleepMinutes', y='AvgSteps')
    plt.title('Average Sleep Minutes vs Average Daily Steps')
    plt.xlabel('Average Sleep Minutes')
    plt.ylabel('Average Daily Steps')
    plt.show()

# 4. Trends over time: Average daily steps over time (all users)
steps_by_date = daily_activity.groupby('ActivityDate')['TotalSteps'].mean()
plt.figure(figsize=(12,6))
steps_by_date.plot()
plt.title('Average Daily Steps Over Time')
plt.xlabel('Date')
plt.ylabel('Average Steps')
plt.show()

# 5. Sedentary vs Active Minutes ratio analysis
daily_activity['ActiveMinutes'] = daily_activity['VeryActiveMinutes'] + daily_activity['FairlyActiveMinutes'] + daily_activity['LightlyActiveMinutes']
daily_activity['SedentaryRatio'] = daily_activity['SedentaryMinutes'] / (daily_activity['ActiveMinutes'] + daily_activity['SedentaryMinutes'])

plt.figure(figsize=(10,6))
sns.histplot(daily_activity['SedentaryRatio'], bins=30, kde=True)
plt.title('Distribution of Sedentary Ratio')
plt.xlabel('Sedentary Ratio')
plt.ylabel('Frequency')
plt.show()

# Save the cleaned and enriched dataset
daily_activity.to_csv('daily_activity_cleaned_enriched.csv', index=False)
print("Cleaned and enriched daily activity data saved.")

```

```python
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyArrowPatch

# Paths
header_image_path = "/mnt/data/0a9a47b1-bb4a-471b-a5ae-4851ba8111b7.png"
output_path = "/mnt/data/cluster_to_campaign_infographic.png"

# Load header image (if available)
try:
    header_img = Image.open(header_image_path)
    has_header = True
except Exception:
    header_img = None
    has_header = False

# Create figure
fig = plt.figure(figsize=(12, 6))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.axis('off')

# If header image exists, place it at the top
if has_header:
    ax_img = fig.add_axes([0.02, 0.78, 0.96, 0.18])
    ax_img.imshow(header_img)
    ax_img.axis('off')

# Coordinates for cluster boxes (left) and campaign boxes (right)
clusters = [
    {"name": "Cluster 1\n(Sedentary)", "xy": (8, 58), "desc": "Low steps\nHigh sedentary minutes\nTarget: small-step goals"},
    {"name": "Cluster 2\n(Moderately Active)", "xy": (8, 33), "desc": "Average steps\nOccasional activity\nTarget: consistency nudges"},
    {"name": "Cluster 3\n(Highly Active)", "xy": (8, 8), "desc": "High steps\nActive minutes\nTarget: advanced insights"}
]

campaigns = [
    {"name": "Campaign A\nSmall Steps Challenge", "xy": (62, 60), "desc": "Micro-goals & badges\nDaily 10–15min prompts"},
    {"name": "Campaign B\nWeekend Warmups", "xy": (62, 35), "desc": "Weekend nudges\nMini-challenges & rewards"},
    {"name": "Campaign C\nAdvanced Insights", "xy": (62, 10), "desc": "Performance metrics\nTraining plans & tips"}
]

# Draw title
ax.text(50, 92, "Cluster → Campaign Mapping", ha='center', va='center', fontsize=18, weight='bold')

# Draw cluster boxes and text
box_w, box_h = 34, 14
for c in clusters:
    x, y = c["xy"]
    rect = Rectangle((x, y), box_w, box_h, fill=False, linewidth=2)
    ax.add_patch(rect)
    ax.text(x + 2, y + box_h - 3, c["name"], fontsize=11, weight='bold', va='top', ha='left')
    ax.text(x + 2, y + box_h - 8, c["desc"], fontsize=9, va='top', ha='left')

# Draw campaign boxes and text
for c in campaigns:
    x, y = c["xy"]
    rect = Rectangle((x, y), box_w, box_h, fill=False, linewidth=2)
    ax.add_patch(rect)
    ax.text(x + 2, y + box_h - 3, c["name"], fontsize=11, weight='bold', va='top', ha='left')
    ax.text(x + 2, y + box_h - 8, c["desc"], fontsize=9, va='top', ha='left')

# Draw arrows mapping clusters to campaigns
arrow_props = dict(arrowstyle='->', linewidth=2)
mappings = [(0, 0), (1, 1), (2, 2)]  # cluster index -> campaign index

for i_cluster, i_campaign in mappings:
    cx, cy = clusters[i_cluster]["xy"]
    px = cx + box_w
    py = cy + box_h / 2
    tx, ty = campaigns[i_campaign]["xy"]
    sx = tx
    sy = ty + box_h / 2
    arrow = FancyArrowPatch((px + 1, py), (sx - 1, sy), connectionstyle="arc3,rad=0.2", **arrow_props)
    ax.add_patch(arrow)

# Footer / legend
ax.text(2, 2, "Notes: Clusters derived from Steps & Sedentary Ratio. Map campaigns to user segments for targeted messaging.", fontsize=9)

# Save and show
plt.savefig(output_path, bbox_inches='tight', dpi=150)
plt.show()
```
